## 2.Line Chart

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=big[big['batsman']=='V Kohli']
```

```python
performance=batsman_data.groupby('season').sum()['batsman_runs']
```

```python
plt.plot(performance.index, performance.values)
```

```python
plt.plot(performance.index, performance.values, color="green")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by Season Performance")
```

```python
plt.plot(performance.index, performance.values, color="green",marker='D')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by Season Performance")
```

```python
batsman_data=big[big['batsman']=='V Kohli']
batsman_data1=big[big['batsman']=='RG Sharma']
performance=batsman_data.groupby('season').sum()['batsman_runs']
performance1=batsman_data1.groupby('season').sum()['batsman_runs']
plt.plot(performance.index, performance.values, color="green",marker='D',label="Virat Kohli")
plt.plot(performance1.index, performance1.values, color="red",marker='+',label="Rohit Sharma")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by Season Performance")
plt.legend()
```

```python
def comparator(batsman1,batsman2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==batsman1]
    batsman_data1=big[big['batsman']==batsman2]
    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(performance.index, performance.values, color="green",marker='D',label=batsman1)
    plt.plot(performance1.index, performance1.values, color="red",marker='+',label=batsman2)
    plt.xlabel('Season')
    plt.ylabel('Runs Scored')
    plt.title("Batsman Season by Season Performance")
    plt.legend()
    
    
```

```python
comparator("MS Dhoni","AB de Villiers")
```
